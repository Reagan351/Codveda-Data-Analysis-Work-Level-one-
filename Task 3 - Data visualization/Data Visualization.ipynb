```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("../Task 1 - Data Cleaning and preprocessing/my_cleaned_data.csv")
df.head()
```

```python
# Creating a bar plot
plt.figure()
sns.countplot(data=df, x="species")
plt.title("Count of Flower Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
# Exporting the bar plot as image
plt.figure()
sns.countplot(data=df, x="species")
plt.title("Count of Flower Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.savefig("bar_plot_species.png")
plt.show()
```

```python
# Create a line chart
plt.figure()
plt.plot(df["sepal_length"])
plt.title("Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
# Exporting the line chart as image so that it can be saved
plt.figure()
plt.plot(df["sepal_length"])
plt.title("Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.savefig("line_chart_sepal_length.png")
plt.show()
```

```python
# Create a scatter plot
plt.figure()
sns.scatterplot(
    data=df,
    x="petal_length",
    y="petal_width",
    hue="species"
)
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend(title="Species")
plt.show()
```

```python
# Exporting the scatter plot as image
plt.figure()
sns.scatterplot(
    data=df,
    x="petal_length",
    y="petal_width",
    hue="species"
)
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend(title="Species")
plt.savefig("scatter_plot_petal.png")
plt.show()
```
