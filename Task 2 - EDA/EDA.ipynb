```python
# Importing required directories
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../Task 1 - Data Cleaning and preprocessing/my_cleaned_data.csv")
df.head()
```

```python
# Calculating Summarry Statistics (mean,median,mode.standard deviation)
df.describe()
```

```python
# Mean
df.mean(numeric_only=True)
```

```python
# Median
df.median(numeric_only=True)
```

```python
# Mode
df.mode()
```

```python
# Standard Deviation
df.std(numeric_only=True)
```

```python
# Visualize data distribution
## Histograms
df.hist(figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
# Boxplots
plt.figure(figsize=(8, 5))
sns.boxplot(data=df.drop(columns=["species"]))
plt.show()
```

```python
# Scatter plot
sns.scatterplot(
    data=df,
    x="sepal_length",
    y="petal_length",
    hue="species"
)
plt.show()
```

```python
# Find correlations between numerical features
correlation_matrix = df.drop(columns=["species"]).corr()
correlation_matrix
```

```python
# Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```
